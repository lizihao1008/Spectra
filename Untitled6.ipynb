```python
from tardis import *
```

```python
zz = tf.Variable(np.random.random(10),trainable=True)
logposterior = (tf.reduce_sum(zz**2))
opt = ScipyOptimizerInterface(logposterior,var_list=[zz],method='L-BFGS-B',options={'maxiter':10,'disp':True})
```

```python
dir(opt)
```

```python
sess =  tf.Session()
sess.run(tf.global_variables_initializer())
opt.minimize(sess)
print(sess.run([zz]))
```
