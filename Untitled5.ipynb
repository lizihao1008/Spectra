```python
class reconstruct_opt():

    """ Optimization class (Should think of a better name...)

        Directly reconstruct flux field without doing gravitational evolution via flowPM.
        
        Inputs
            ------
            
            Data and Evolution Classes
            
                uni: Gets evolution operators from this class as well as fiducial PK
                lya: Class for lyman alpha observations
                ipklin: (should just get this from Universe object...)
            
            Configuration Settings
            
                anneal_steps: optimization progresses with successive smoothing steps
                maxiter: maximum number of iterations per anneal step 
                gtol: tolerance for convergence criteria
                k: An array of shape 1*4 containing weighting factors for different terms in the loss function. 
                clip: bool. If ``True``, the loss  function will contain clip term to penaliza flux values 
                      over one, which usually outperforms standard loss function. The default is ``True ``
            Outputs
            -------
                
                tau_recon : Reconstructed optical depth field
                recon_flux : flux along individual LOS determined by lya
                opt: optimizer output, includes hessian and jacobean information for "something"
            
            Sample Usage
            -------
            
                q = reconstruct_opt(uni,lya=desi,
                                    maxiter=[5,5,5,10],
                                    anneal_steps=[2.5,1.5,1,0.7],
                                    k = [1,2,1,0.8])
                q.initmodel() #initializes graph
                q.run_model() #creates TF session and runs optimization

    """

    def __init__(self,uni,lya,ipklin = ipklin,maxiter=[50,50,60,30],
                 anneal_steps=[2.5,1.5,1,0.7],k = [1,2,1,0.8],gtol=1e-5/512
                ):
        
        
        self.maxiter = maxiter
        self.gtol = gtol
        self.anneal_steps = anneal_steps
        
        self.uni = uni
        self.k = k
        self.lya = lya
        self.ipklin = ipklin
        self.bound = bound

    def __init__prior(self,nc,bs):

        kvec = fftk((nc,nc,nc), symmetric=False)
        self.kmesh = sum((np.array(kvec,dtype=object))**2)**0.5
        self.pkmesh = self.ipklin(self.kmesh)

    def initmodel(self,random_seed=None):

        tf.compat.v1.reset_default_graph()

        np.random.seed(random_seed)
        tf.compat.v1.set_random_seed(random_seed)

        nc = self.uni.nc
        bs = self.uni.bs
        self.__init__prior(nc,bs)

        self.initial=tf.Variable(np.zeros(shape=(1,nc,nc,nc)),dtype='float32',trainable=True)
        self.RSD_tau = tf.multiply(self.initial,1)
        self.Rs = tf.placeholder(tf.float64, shape = [1], name='smoothing')

        sum_op = reduce_sum_det
        Rsm = tf.multiply(self.Rs, 1)

        flux = tf.exp(-self.RSD_tau)
        flux_s = smoothing(flux,Rsm,self.kmesh,nc)
        self.yy_lya = tf.placeholder(tf.float32, [self.lya.skewers_fin.shape[0]]) #lya flux
        self.flux_val = flowpm.cic_readout(flux_s,np.array(self.lya.skewers_fin,dtype=np.float32).reshape(1,-1,3))



        def chi(inputs, data, inverse_covar):
    
            residual = tf.subtract(inputs, data)
            a = tf.sparse.sparse_dense_matmul(inverse_covar, residual,adjoint_b=True)
            value = tf.matmul(residual,a)
            return value

        def overone(flux):

            flux = K.clip(flux,1,None)
            return(tf.reduce_sum(flux))

        def nodes(flux):

            flux = K.clip(flux,0,0.4)
            return(tf.reduce_sum(flux))
        #The final loss functon we want to minimize
        
        self.obj_fun = self.k[0]*(chi(self.flux_val,self.yy_lya,self.lya.inverse_covar)+self.k[1]*sum_op(tf.square(tf.subtract(flux_s,flux)))
                    +self.k[2]*overone(flux)+self.k[3]*nodes(flux))
        


    def run_model(self):

        sess = tf.Session()
        sess.run(tf.compat.v1.global_variables_initializer())

        time_start = time.time()
        for nn,r0 in enumerate(self.anneal_steps):
            self.opt = ScipyOptimizerInterface(self.obj_fun, var_list=[self.initial], method='L-BFGS-B',
                                               options={'maxiter': self.maxiter[nn], 'gtol':self.gtol,'disp': True})
 
            self.z = self.opt.minimize(sess, {self.yy_lya:self.lya.vals_observed,self.Rs:[r0]})
        time_end = time.time()
        print('Total run time %f'%(time_end-time_start))
        self.recon_RSD_tau,self.recon_flux = sess.run([self.RSD_tau,self.flux_val],{self.yy_lya:self.lya.vals_observed,self.Rs:[0]})

        return self.recon_RSD_tau,self.recon_flux
```
